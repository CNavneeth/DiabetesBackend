# Hypertension

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('Hypertension-risk-model-main.csv')
data.head()
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
data['cigsPerDay'].fillna(data.cigsPerDay.mode()[0],inplace=True)
data
```

```python
data['BMI'].fillna(data['BMI'].mean(),inplace=True)
data['glucose'].fillna(data['glucose'].mean(),inplace=True)
```

```python

sns.kdeplot(data['BPMeds'].dropna(), shade=True)
plt.title('KDE Plot')
plt.show()
```

```python
data['BPMeds'].fillna(method='ffill',inplace=True)
data['totChol'].fillna(data['totChol'].median(),inplace=True)
data['heartRate'].fillna(data['heartRate'].median(),inplace=True)
```

```python
data.isnull().sum()
```

```python
sns.countplot(x='male', hue='Risk', data=data)
plt.xticks(ticks=[0,1], labels=['Female','Male'])
plt.legend(labels=['Female','Male'])
plt.show()
```

```python

sns.histplot(data=data[data['Risk']==1], x=data.male)
plt.xticks(ticks=[0,1], labels=['Female','Male'])
```

```python
plt.pie(data['Risk'].value_counts(),labels=['No Disease','Disease'],autopct='%1.1f%%',startangle=60)
plt.legend()#,labels=['Risk','Risk-Free'])
plt.show()
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X=data.drop('Risk',axis=1)
y=data['Risk']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y, test_size=0.20, random_state=40)
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
Reg=LogisticRegression(random_state=40)
```

```python
Reg.fit(X_train,y_train)
```

```python
y_pred=Reg.predict(X_test)
```

```python
reg_score=accuracy_score(y_test,y_pred)
```

## Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier,plot_tree
```

```python
tree=DecisionTreeClassifier(random_state=30, max_depth=4)
```

```python
tree.fit(X_train,y_train)
```

```python
y1_pred=tree.predict(X_test)
```

```python
tree_score=accuracy_score(y_test,y1_pred)
```

```python
plt.figure(figsize=(30,30))
plot_tree(tree,
          
          feature_names=X.columns,  # If X is a pandas DataFrame, you can use column names
          class_names=['Class 0', 'Class 1'],  # Adjust this if your target labels are different
          rounded=True,
          proportion=True,# Rounded corners for a more professional look
          fontsize=16,
         label='none')
plt.show()
```

## Compare Accuracy

```python
accuracy_table=pd.DataFrame({'Logistic Regression': [reg_score],'Decision Tree': [tree_score]},index=['accuracy'])
```
